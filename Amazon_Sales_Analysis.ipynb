```python
# By using this "for-loop" we can easily get data labels in seaborn, which in matplotlib is bit tough.

sizeCount = sns.countplot(x = 'Size', data = df) 
for bars in sizeCount.containers:
    sizeCount.bar_label(bars)
```

Most of the People Buys M-Size

```python
#  hue parameter is used to group and differentiate the bars based on a categorical variable.
#  It adds color to distinguish different categories within the same x variable.

#  hue = 'Status' → Each bar is split into different colors based on the Status column.
```

```python
plt.figure(figsize = (10,6))
st = sns.countplot(data = df , x = 'Courier Status' , hue = 'Status')

for bars in st.containers:
    st.bar_label(bars)

plt.show()
    
```

```python
# df['Category'] = df['Category'].astype(str)
cate = df['Category']
plt.figure(figsize = (10,6))
plt.hist(cate , bins = 15 , edgecolor = 'Red')
plt.xticks(rotation = 0)
plt.show()
```

```python

Check_B2B = df['B2B'].value_counts()

plt.pie(Check_B2B, labels=Check_B2B.index, autopct="%1.1f%%", wedgeprops={'edgecolor': 'white'})

# Add a white circle at the center to make it a donut chart
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("B2B Distribution")
plt.show()
```

```python
x_data = df['Category']
y_data = df['Size']

plt.scatter(x_data , y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Available Size')
```

```python
# Get the top 10 states by order count
top_states = df['ship-state'].value_counts().nlargest(10).index

# Filter the dataset to include only these top 10 states
df_top_states = df[df['ship-state'].isin(top_states)]

plt.figure(figsize=(12, 8))
sns.countplot(data=df_top_states, x="ship-state", order=top_states)

plt.xlabel('State')
plt.ylabel('Order Count')
plt.title('Top 10 States by Order Count')
plt.xticks(rotation = 45)
plt.show()
```
